### What is the distribution of $X$ (mean and dispersion)?

**The Model: $X = AF + G$ with**

$F \sim \mathcal N(0, I_2)$, $G \sim \mathcal N\!\left(0,\,\mathrm{diag}\!\left(\tfrac49,\tfrac79,\tfrac49\right)\right)$, and $F \perp G$.

**The Mean:**

$$
\mathbb E[X]=A\,\mathbb E[F]+\mathbb E[G]=0.
$$

**The Covariance (dispersion):**

$$
\Sigma_X=\operatorname{Cov}(X)=A\,\operatorname{Cov}(F)\,A^\top+\operatorname{Cov}(G)
        =A A^\top+\mathrm{diag}\!\left(\tfrac49,\tfrac79,\tfrac49\right).
$$

With

$$
A=\begin{bmatrix}
\tfrac23 & \tfrac13\\[2pt]
\tfrac13 & \tfrac13\\[2pt]
-\tfrac13 & \tfrac23
\end{bmatrix},
\quad
A A^\top=\begin{bmatrix}
\tfrac59 & \tfrac13 & 0\\[2pt]
\tfrac13 & \tfrac29 & \tfrac19\\[2pt]
0 & \tfrac19 & \tfrac59
\end{bmatrix}.
$$

Add the uniques:

$$
\Sigma_X
=\begin{bmatrix}

### What are the squared multiple correlations $\rho_{1|23}^2, \, \rho_{2|13}^2$ and $\rho_{3|12}^2$ between an X-coordinate and the two other X-coordinates?

We want $\rho^2_{1|23}, \rho^2_{2|13}, \rho^2_{3|12}$. These are the squared multiple correlations between one coordinate of $X$ and the two other coordinates. For a Gaussian vector $X \sim \mathcal N(0,\Sigma)$, the squared multiple correlation of $X_i$ with the others is

$$
R_i^2 = 1 - \frac{\operatorname{Var}(X_i \mid X_{-i})}{\operatorname{Var}(X_i)}.
$$

From part (2), we had

$$
\Sigma =
\begin{bmatrix}
1 & \tfrac{1}{3} & 0 \\
\tfrac{1}{3} & 1 & \tfrac{1}{9} \\
0 & \tfrac{1}{9} & 1
\end{bmatrix}.
$$

Each diagonal element is 1, so $\operatorname{Var}(X_i)=1$.

For a multivariate normal, the conditional variance is given by the precision matrix $K = \Sigma^{-1}$:

$$
\operatorname{Var}(X_i \mid X_{-i}) = \frac{1}{K_{ii}}.
$$

So

$$
R_i^2 = 1 - \frac{1}{K_{ii}}.
$$

$$
K = \Sigma^{-1} \;=\; \frac{1}{71}
\begin{bmatrix}
80 & -27 & 3 \\
-27 & 81 & -9 \\
3 & -9 & 72
\end{bmatrix}.
$$

So the diagonal elements are

$$
K_{11} = \tfrac{80}{71}, \quad K_{22} = \tfrac{81}{71}, \quad K_{33} = \tfrac{72}{71}.
$$

```r
library(MASS)

Sigma <- matrix(c(1, 1/3, 0,
                  1/3, 1, 1/9,
                  0, 1/9, 1), 3, 3, byrow = TRUE)

K <- solve(Sigma)
R2 <- 1 - 1/diag(K)

fractions(K)   # precision matrix as fractions
fractions(R2)  # squared multiple correlations
```

\end{aligned}
$$
