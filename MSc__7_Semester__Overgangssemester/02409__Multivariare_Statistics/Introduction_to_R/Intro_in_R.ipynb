## slide 16: functions in R

```r
# a simple function
f<-function(x){x^2}; f(2)
```

```r
# a function of two variables
f<-function(x,pow){x^pow}; f(2,2)
```

```r
# a function with a default value.
f<-function(x,pow=2){x^pow}; f(2,2); f(2);f(2,3)
```
