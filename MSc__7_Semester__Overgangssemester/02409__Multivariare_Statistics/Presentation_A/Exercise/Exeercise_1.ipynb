### b) Determine the axes in the contour ellipsoid for the probability density function of X

To determine the axes in the contour ellipsoid for the probability density function of $X$, we need to find the eigenvectors to get the axes of the ellipsoid.

The **principal axes** of those ellipsoids are the **eigenvectors of $\Sigma$**, and the **semi-axis lengths** are proportional to $\sqrt{\lambda_i}$, where $\lambda_i$ are the eigenvalues of $\Sigma$.

We already have the eigenvalues, so now we find the eigenvectors

```python
eigs_full = D.eigenvects()
eigs_full
```

The eigenvectors is:

$$
v_1 =
\begin{bmatrix}-1 \\ 1 \\ 0\end{bmatrix}, \quad
v_1 =
\begin{bmatrix}-1 \\ 0 \\ 1\end{bmatrix}, \quad
v_2 =
\begin{bmatrix}1 \\ 1 \\ 1\end{bmatrix}.
$$

And get the Eigenpairs:

Eigenvalue  
$$
\lambda_1 = 1 - \rho \quad (\text{multiplicity 2}),
$$  
with eigenvectors spanning the plane orthogonal to $(1,1,1)$, e.g.  
$$
v_1=\begin{bmatrix}-1\\1\\0\end{bmatrix},\quad 
v_1=\begin{bmatrix}-1\\0\\1\end{bmatrix}.
$$

Eigenvalue  
$$
\lambda_2 = 1 + 2\rho \quad (\text{multiplicity 1}),
$$  
with eigenvector  
$$
v_2=\begin{bmatrix}1\\1\\1\end{bmatrix}.
$$

Now we need to normalize them to unit length, so that we obtain an orthonormal basis 
for the ellipsoid axes. The directions are unchanged, but working with normalized 
eigenvectors makes the geometry and axis lengths easier to describe.

```python
# Define normalized versions for convenience
v1 = Matrix([-1, 1, 0]) / sqrt(2)
v2 = Matrix([-1, 0, 1]) / sqrt(2)
v3 = Matrix([1, 1, 1]) / sqrt(3)

v1, v2, v3
```

This give us:
$$
\hat v_1 = \tfrac{\sqrt{2}}{2}\begin{bmatrix}-1\\1\\0\end{bmatrix},\quad
\hat v_1 = \tfrac{\sqrt{2}}{2}\begin{bmatrix}-1\\0\\1\end{bmatrix},\quad
\hat v_2 = \tfrac{\sqrt{3}}{3}\begin{bmatrix}1\\1\\1\end{bmatrix}
$$

For $v_1$ and $v_1$, the eigenvalue is $1-\rho$.  

Since we have an eigenvalue with multiplicitet 2, there are to versions of $v_1$. To continue, we can onlu pick one of them which is: 
$$
\hat v_1 = \tfrac{\sqrt{2}}{2}\begin{bmatrix}-1\\1\\0\end{bmatrix},\quad
$$

For $v_2$, the eigenvalue is $1+2\rho$ we have the eigenvecctor:
$$
\hat v_2 = \tfrac{\sqrt{3}}{3}\begin{bmatrix}1\\1\\1\end{bmatrix}
$$  

Now we must find the last eigenvector, which is a linear combination of the two above, because for one of the eigenvalues, we have multiplicity of 2, meaning that the space of 3 is not fully spanned. Hence, we must do this extra step.

Recalling our two egienvectors, we know that $v_3$ must satisfy:
$$
-x_1 + x_2 + 0 = 0
$$
$$
x_1 + x_2 + x_3 = 0
$$
We solve this in sympy below.

```python
x1, x2, x3 = symbols('x1 x2 x3')

eq1 = Eq(-x1 + x2 + 0, 0)
eq2 = Eq(x1 + x2 + x3, 0)

solve((eq1, eq2), (x1, x2, x3))
```

Forming a vector out of $x_3$ gives us:
$$
v_3 = x_3 \begin{bmatrix} -\frac{1}{2} \\ -\frac{1}{2} \\ 1 \end{bmatrix}
$$
We scale this with $2$:
$$
v_3 = x_3 \begin{bmatrix} -1 \\ -1 \\ 2 \end{bmatrix}
$$
Now we will normalize in Python.

```python
v3 = Matrix([[-1, -1, 2]])

v3_norm = v3 / v3.norm()

v3_norm
```

The three normalized eigenvectors are thus the axes directions of the ellipsoid, and the eigenvalues are the lengths of the semi-axes:
$$
v_1 = \frac{\sqrt{2}}{2} \begin{bmatrix} -1 \\ 1 \\ 0 \end{bmatrix}, \: v_2 = \frac{\sqrt{3}}{3} \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}, \: v_3 = \frac{\sqrt{6}}{6} \begin{bmatrix} -1 \\ -1 \\ 2 \end{bmatrix}
$$

To fin the conditional mean of $E\left(\begin{bmatrix} U \\ V \end{bmatrix} | Z = z\right)$, we use the formula from slide 71 given as:
$$
E(X_1|X_2=x_2)=\mu_1 +\Sigma_{11}\Sigma_{22}^{-1}(x_2-\mu_2)
$$

We have to do some partitioning of vector $\begin{bmatrix} X \\ Y \\ Z \end{bmatrix}$, where we then will have $\begin{bmatrix} X \\ Y \end{bmatrix}$ and $\begin{bmatrix} Z \end{bmatrix}$, and hence, we must also partition the disperision matrix accordingly:
$$
\Sigma = \begin{bmatrix} \Sigma_{11} & \Sigma_{12} \\ \Sigma_{21} & \Sigma_{22} \end{bmatrix}
$$

Since, we have a two dimensional vector on the left side of the equation, then our $\Sigma_{11}$ must also be the two dimensional, in order to span both $X$ and $Y$.
With:
$$
\Sigma_{11} = \begin{bmatrix} 1 & \rho \\ \rho & 1 \end{bmatrix}
$$
$$
\Sigma_{12} = \begin{bmatrix} \rho \\ \rho \end{bmatrix}
$$
$$
\Sigma_{21} = \begin{bmatrix} \rho & \rho \end{bmatrix}
$$
$$
\Sigma_{22} = [1]
$$
$$
\mu_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$
$$
\mu_2 = 1
$$
Moreover, we have that $\mu_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$ and $\mu_2 = [1]$. We put everything in Python.

```python
sigma_12 = Matrix([[rho],[rho]])
sigma_22 = Matrix([1])
mu_1 = Matrix([[1],[1]])
mu_2 = Matrix([1])

z = symbols("z")

x_2 = Matrix([z])

mu_1+sigma_12*sigma_22.inv()*(x_2-mu_2)
```
