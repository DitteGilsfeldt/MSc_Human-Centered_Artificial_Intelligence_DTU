### a) For which values of ρ is this a valid dispersion matrix? (Hint: Both dispersion and correlation matrices must be positive semi-definit)

```python
from sympy import *
init_printing()

rho = symbols("ρ")
D = Matrix([[1, rho, rho],
            [rho, 1, rho],
            [rho, rho, 1]])
D
```

I'll start checking if the matrices is positive semi-definit, by start finding the eigenvalues. It is not necessary to find the eigenvectors to see if the matrix is positive semi-definite.

```python
eigs = D.eigenvals()
eigs
```

**Eigenvalue:**

$
\lambda_1 = 1 - \rho
$\
with **multiplicity 2**.  


<!-- A basis for its eigenspace is\
$

\begin{bmatrix}

\end{bmatrix},
\qquad
v_{2} =
\begin{bmatrix}

\end{bmatrix}.
$ -->

**Eigenvalue:**

$
\lambda_2 = 1 + 2\rho
$\
with **multiplicity 1**.  


<!-- Its eigenvector is\
$

$ -->


Now, I have finding the eigenvalues, and then I need to see if $\lambda \geq 0$. This will be finding by isolate $\rho$.

```python
Lambda1 = 1 - rho
Lambda2 = 1 + 2*rho

sols1 = solve(Lambda1, rho)
sols2 = solve(Lambda2, rho)

print(sols1, sols2)
```

This means that the range of $\rho$  is:\
$ -\frac{1}{2}\leq \rho \leq 1$
