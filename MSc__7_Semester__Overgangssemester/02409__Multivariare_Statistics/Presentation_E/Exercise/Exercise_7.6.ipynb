**Canonical Correlation Analysis on Beef Characterization**\
We consider the data from Exercise 7.3 and do now want to analyze the relation between the panel
assessments and the physical/chemical measurements on the meat by means of Canonical Correlation
Analyses of the data. A sample program for this is:

```r
Exx = as.matrix(Beef[1:11,1:11])
Eyx = as.matrix(Beef[12:18,1:11])
Exy = as.matrix(Beef[1:11,12:18])
Eyy = as.matrix(Beef[12:18,12:18])
invExx = solve(Exx)
invEyy = solve(Eyy)

#Calculating the Canonical correlations:
Cancorr = geigen(Eyx%*%invExx%*%Exy,Eyy,symmetric = TRUE)
values = sort(Cancorr$values,decreasing = TRUE)

# E is the residual variation after having predicted Y by means of X
H = Eyx%*%invExx%*%Exy
E = Eyy - Eyx%*%invExx%*%Exy
invE = solve(E)
Ev <- eigen(invE%*%H)
var = Ev$values

# Eigenvalues, Proportion and Cumulative proportion of Variance:
varPC <- var/sum(var)
cumu = c(1:7)
for (i in 1:7){
  cumu[i] = sum(varPC[1:i])
}
results <- data.frame("CanCor" = sqrt(values),"Squared CanCor" = values,"eigenvalues"=var,"proportion"=varPC,
                      "cumulative" = cumu)
print("Table with information about the Canonical Correlations:")
results
```

```r
## Calculating p-values using F-approximations with Wilks test:
n = 79
p = length(Exx[,1])
q = length(Eyy[,1])


HypTest <- p.asym(results$CanCor,n,p,q,tstat="Wilks")
```

```r
results_all <- data.frame(results,HypTest)
print("Table with information about the Canonical Correlations similar to the output in SAS:")
results_all
```

## 1. How many Canonical Correlation Coefficients are significantly different from 0?

**What are Canonical Correlation Coefficients?**\
Canonical Correlation Analysis (CCA) finds linear combinations of two sets of variables (say X and Y) such that the correlations between them are maximized.
You get multiple canonical correlation coefficients, but not all of them are statistically significant.

[(See slide from week E, pages 39 and 40)](../Slide_ect/Presentation%20E%20ANST%202025.pdf)

Assume multivariate normality. Then each canonical pair $((a_p^\top Y,; b_p^\top X))$ is bivariate normal with unit variances and correlation $(\rho_p)$.

Test $(H_0:\rho_k=\cdots=\rho_s=0)$ sequentially with Wilks’ $(\Lambda_k=\prod_{i=k}^s(1-r_i^2))$.

Use Bartlett’s chi-square: $(\chi^2 = -\big[n-1-(p+q+1)/2\big]\ln\Lambda_k)$, $df (= m(p+q-m+1))$, $(m=s-k+1)$.

Stop at the first non-significant test. That count = # significant canonical correlations.

```r
# Significance test
p.asym(results$CanCor, n, p, q, tstat = "Wilks")
```

```r
# ---- Count how many are significant (stop at first non-significant) ----
alpha <- 0.05
wilks <- HypTest
rhos <- results$CanCor
first_nonsig <- which(wilks$p.value > alpha)[1]
num_sig <- if (is.na(first_nonsig)) nrow(wilks) else first_nonsig - 1

cat("Canonical correlations:\n"); print(round(rhos, 4))
cat("\nWilks table:\n"); print(wilks)
cat("\n# of canonical correlations significantly > 0 (alpha =", alpha, "): ", num_sig, "\n", sep = "")
```
