* Explain in your own words what aspect of language _lexical diversity_ describes. 

The *lexical diversity* is a plot that describes how many words that appears in the text from the beginning. 


It is one thing to automatically detect that a particular word occurs in a text, and to display some words that appear in the same context. However, we can also determine the location of a word in the text: how many words from the beginning it appears. This positional information can be displayed using a dispersion plot. Each stripe represents an instance of a word, and each row represents the entire text. In 1.2 we see some striking patterns of word usage over the last 220 years (in an artificial text constructed by joining the texts of the Inaugural Address Corpus end-to-end). You can produce this plot as shown below. You might like to try more words (e.g., liberty, constitution), and different texts. Can you predict the dispersion of a word before you view it? As before, take care to get the quotes, commas, brackets and parentheses exactly right.

**Lexical diversity** is a measure of how varied the vocabulary is in a piece of writing or speech.

In simple terms, it tells you how many different words are used compared to the total number of words.

* If a text repeats the same words often, it has **low lexical diversity**.
* If a text uses many different words without much repetition, it has **high lexical diversity**.

So, lexical diversity shows how rich or varied someones language is.

* What is a bigram? How does it relate to `collocations`. Explain in your own words.

1. What is a bigram?

A bigram is a pair of two consecutive words in a text.

For example, in the sentence:\
The quick brown fox jumps, the bigrams are:
* (The, quick)
* (quick, brown)
* (brown, fox)
* (fox, jumps)

So, a bigram captures the relationship between neighboring words  basically, what words tend to appear together.

2. What are collocations?

Collocations are combinations of words that frequently appear together more often than would happen by chance.

Examples:
* strong coffee
* make a decision
* fast food

These pairs sound natural together to native speakers because they occur so commonly.

3. How bigrams relate to collocations

Bigrams are the building blocks used to find collocations.

When you extract all bigrams from a text and then measure which ones occur unusually often, those frequent pairs are your collocations.

In short:
* All collocations are bigrams,
* but not all bigrams are collocations  only the meaningful, statistically frequent ones.

* What has happened to the usage of these words over time?

````python
from nltk.corpus import state_union
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

men = []
women = []
ppl = []
year = []


for id in state_union.fileids():
    text = state_union.words(id)
    year.append(id[0:4])
    men.append(text.count('men'))
    women.append(text.count('women'))
    ppl.append(text.count('people'))

plt.figure(figsize=(12, 6))
plt.plot(year, men, label='Men')
plt.plot(year, women, label='Women')
plt.plot(year, ppl, label='People')
plt.gca().xaxis.set_major_locator(MaxNLocator(10))
plt.legend()
plt.show()
````

````python
# Make plot of men and women over time in each text:
men_counts = []
women_counts = []
people_counts = []

files = state_union.fileids()
years = [int(file[:4]) for file in files]

for file in files:
    words = [w.lower() for w in state_union.words(file)]
    men_counts.append(words.count("men"))
    women_counts.append(words.count("women"))
    people_counts.append(words.count("people"))

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(years, men_counts, label="Men")
plt.plot(years, women_counts, label="Women")
plt.title("Mentions of 'men' and 'women' in through 1945-2006")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.show()
````

16. Write a program to generate a table of lexical diversity scores (i.e. token/type ratios), as we saw in 1.1. Include the full set of Brown Corpus genres (nltk.corpus.brown.categories()). Which genre has the lowest diversity (greatest number of tokens per type)? Is this what you would have expected?

<img src="SymPyBilleder/2025-10-08-13-17-06.png" width="550">

````python
# Write a program to generate a table of lexical diversity scores (i.e. token/type ratios). 

# Function to calculate lexical diversity for a given genre
def lexical_diversity(text):
    return len(set(text)) / len(text)

# Loop through all genres in the Brown corpus
for genre in brown.categories():
    words = brown.words(categories=genre)
    diversity = lexical_diversity(words)
    print(f"{genre:20s} | Lexical Diversity: {diversity:.3f}")
````

* Which genre has the lowest diversity (greatest number of tokens per type)?

learned has the lowest diversity. That means it repeats words more often, it has many tokens per unique word.

* Is this what you would have expected?

Partly yes, partly no.

Why it makes sense:\
Academic or technical writing tends to repeat specialized vocabulary.\
Terms and key concepts (e.g., theory, method, data, results) occur many times, which lowers lexical diversity.\
These texts focus on precision and consistency, not variety of word choice.

Why it might seem unexpected:\
You might expect academic writing to use a wider vocabulary (it feels more complex).\
But complex here means technical, not varied.\
Scholarly texts often reuse a limited set of terms consistently, so their diversity score goes down.

17. Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

````python
from nltk.corpus import stopwords

def top_50_words(text):
    """Return the 50 most frequent non-stopword words in a text."""
    
    # Convert all words to lowercase
    words = [w.lower() for w in text if w.isalpha()]  # keep only alphabetic words
    
    # Get English stopwords
    stop_words = set(stopwords.words('english'))
    
    # Remove stopwords
    filtered_words = [w for w in words if w not in stop_words]
    
    # Create frequency distribution
    freq_dist = FreqDist(filtered_words)
    
    # Return 50 most common
    return freq_dist.most_common(50)
````

18. Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

````python
def top_50_bigrams(text):
    """Print the 50 most frequent bigrams excluding stopwords."""
    
    # Keep only alphabetic words and make lowercase
    words = [w.lower() for w in text if w.isalpha()]
    
    # Define stopwords
    stop_words = set(stopwords.words('english'))
    
    # Filter out stopwords
    filtered_words = [w for w in words if w not in stop_words]
    
    # Create bigrams (pairs of consecutive words)
    bigram_list = list(bigrams(filtered_words))
    
    # Count bigram frequencies
    freq_dist = FreqDist(bigram_list)
    
    # Print the 50 most common bigrams
    for pair, freq in freq_dist.most_common(50):
        print(f"{pair}: {freq}")
````
